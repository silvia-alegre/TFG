```python
from evaluate import load

bleu = load('bleu')
comet = load('comet')
```

```python
import pandas as pd

data = pd.read_csv('llama3_results_alpha_128_r_16.csv', index_col = [0])   #

source = [i.lower() for i in data['lemmas']]

prediction = []
for i in data['generated phrases']:  #generated phrases
  try:
    app = i.lower()
    app = app.split('\n')[0].strip('.')
  except:
    app = ''
  prediction.append(app)

#print(prediction)

real = [i.lower().strip('.') for i in data['real phrases']]   # real phrases

bleu_score = bleu.compute(predictions=prediction, references=real)
comet_score = comet.compute(predictions = prediction, references = real, sources = source)

print('BLEU SCORE: ', bleu_score)
print('COMET SCORE: ', comet_score)
```
